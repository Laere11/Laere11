Below is a simple example using matplotlib in Python that:

Prompts you for a 3×3 rotation matrix (row by row).
Draws the original (untransformed) axes in 3D.
Draws the new (transformed) axes by multiplying your rotation matrix with each of the three unit axes vectors.

How It Works
Original Axes: We define x_orig = [1,0,0], y_orig = [0,1,0], and z_orig = [0,0,1].
Transformed Axes: Multiplying each unit vector by your entered rotation matrix R gives the new directions in the old/original coordinate system.
Plotting: We create a single 3D figure, draw each axis as an arrow from the origin to the tip of the axis vector, and label them so you can see how they’ve changed.

```python
import numpy as np
import matplotlib.pyplot as plt

def show_axes_transformation():
    """
    Prompts the user to input a 3x3 rotation matrix, then visualizes:
      1) The original X, Y, Z axes
      2) The new axes after applying the rotation
    """
    # Prompt the user for the 3x3 matrix
    R = []
    print("Enter the 3x3 rotation matrix row by row:")
    for i in range(3):
        row_input = input(f"  Row {i+1} (space-separated): ")
        row_values = list(map(float, row_input.split()))
        R.append(row_values)
    R = np.array(R)

    # Original axis unit vectors
    x_orig = np.array([1, 0, 0])
    y_orig = np.array([0, 1, 0])
    z_orig = np.array([0, 0, 1])

    # New axis vectors = R * original_unit_vectors
    x_new = R @ x_orig
    y_new = R @ y_orig
    z_new = R @ z_orig

    # Create a 3D figure
    fig = plt.figure()
    ax = fig.add_subplot(111, projection='3d')

    # Draw original axes (solid arrows)
    ax.quiver(0, 0, 0, x_orig[0], x_orig[1], x_orig[2], color='r')
    ax.text(x_orig[0], x_orig[1], x_orig[2], "X_orig", color='r')

    ax.quiver(0, 0, 0, y_orig[0], y_orig[1], y_orig[2], color='g')
    ax.text(y_orig[0], y_orig[1], y_orig[2], "Y_orig", color='g')

    ax.quiver(0, 0, 0, z_orig[0], z_orig[1], z_orig[2], color='b')
    ax.text(z_orig[0], z_orig[1], z_orig[2], "Z_orig", color='b')

    # Draw new axes (dashed arrows) to distinguish them
    ax.quiver(0, 0, 0, x_new[0], x_new[1], x_new[2], color='r', linestyle='dashed')
    ax.text(x_new[0], x_new[1], x_new[2], "X_new", color='r')

    ax.quiver(0, 0, 0, y_new[0], y_new[1], y_new[2], color='g', linestyle='dashed')
    ax.text(y_new[0], y_new[1], y_new[2], "Y_new", color='g')

    ax.quiver(0, 0, 0, z_new[0], z_new[1], z_new[2], color='b', linestyle='dashed')
    ax.text(z_new[0], z_new[1], z_new[2], "Z_new", color='b')

    # Some sensible limits so everything is visible
    ax.set_xlim([-1.5, 1.5])
    ax.set_ylim([-1.5, 1.5])
    ax.set_zlim([-1.5, 1.5])

    ax.set_xlabel("X")
    ax.set_ylabel("Y")
    ax.set_zlabel("Z")
    plt.title("Axis Transformation Visualization")

    plt.show()

if __name__ == "__main__":
    show_axes_transformation()
```

Below is an updated version of the code that places original axes in one 3D subplot and the new (transformed) axes in a separate 3D subplot. This way, the new axes are no longer overlapping or hidden by the old ones.

How It Works
User Input: You’re prompted to enter three rows of a 3×3 matrix.
Compute New Axes: We multiply the original unit vectors (1,0,0), (0,1,0), (0,0,1) by the given rotation matrix to find their new orientations in the original frame.
Plotting:
Subplot 1 shows only the original axes.
Subplot 2 shows only the new (transformed) axes.
This keeps them visually separate.










```python
import numpy as np
import matplotlib.pyplot as plt

def show_axes_transformation_separate():
    """
    Prompts the user to input a 3x3 rotation matrix, then visualizes:
      1) The original X, Y, Z axes in one 3D subplot
      2) The new axes (after applying the rotation) in a second 3D subplot
    """

    # Prompt the user for the 3x3 matrix
    R = []
    print("Enter the 3x3 rotation matrix row by row:")
    for i in range(3):
        row_input = input(f"  Row {i+1} (space-separated): ")
        row_values = list(map(float, row_input.split()))
        R.append(row_values)
    R = np.array(R)

    # Original axis unit vectors
    x_orig = np.array([1, 0, 0])
    y_orig = np.array([0, 1, 0])
    z_orig = np.array([0, 0, 1])

    # New axis vectors = R * original_unit_vectors
    x_new = R @ x_orig
    y_new = R @ y_orig
    z_new = R @ z_orig

    # Create a figure with 2 subplots side by side
    fig = plt.figure(figsize=(12, 6))
    ax1 = fig.add_subplot(121, projection='3d')
    ax2 = fig.add_subplot(122, projection='3d')

    # ----------------------------
    # Subplot 1: Original axes
    # ----------------------------
    ax1.quiver(0, 0, 0, x_orig[0], x_orig[1], x_orig[2], color='r')
    ax1.text(x_orig[0], x_orig[1], x_orig[2], "X_orig", color='r')

    ax1.quiver(0, 0, 0, y_orig[0], y_orig[1], y_orig[2], color='g')
    ax1.text(y_orig[0], y_orig[1], y_orig[2], "Y_orig", color='g')

    ax1.quiver(0, 0, 0, z_orig[0], z_orig[1], z_orig[2], color='b')
    ax1.text(z_orig[0], z_orig[1], z_orig[2], "Z_orig", color='b')

    ax1.set_xlim([-1.5, 1.5])
    ax1.set_ylim([-1.5, 1.5])
    ax1.set_zlim([-1.5, 1.5])
    ax1.set_xlabel("X")
    ax1.set_ylabel("Y")
    ax1.set_zlabel("Z")
    ax1.set_title("Original Axes")

    # ----------------------------
    # Subplot 2: New (transformed) axes
    # ----------------------------
    ax2.quiver(0, 0, 0, x_new[0], x_new[1], x_new[2], color='r')
    ax2.text(x_new[0], x_new[1], x_new[2], "X_new", color='r')

    ax2.quiver(0, 0, 0, y_new[0], y_new[1], y_new[2], color='g')
    ax2.text(y_new[0], y_new[1], y_new[2], "Y_new", color='g')

    ax2.quiver(0, 0, 0, z_new[0], z_new[1], z_new[2], color='b')
    ax2.text(z_new[0], z_new[1], z_new[2], "Z_new", color='b')

    ax2.set_xlim([-1.5, 1.5])
    ax2.set_ylim([-1.5, 1.5])
    ax2.set_zlim([-1.5, 1.5])
    ax2.set_xlabel("X")
    ax2.set_ylabel("Y")
    ax2.set_zlabel("Z")
    ax2.set_title("New Axes (Transformed)")

    plt.tight_layout()
    plt.show()

if __name__ == "__main__":
    show_axes_transformation_separate()
```
