This plot show the Battery BMS slave board cell average voltage for the Ford Focus EV Auxillary pak.  The battery has ungone significant balancing over the weekend to narrow the cell variations.  Primarly the low cells were cell #13 and cell #24 on the slave board 1, which took as much as 8Ah of charge in to increase from about 4.00V to 4.07V.  However several other cells in the slave board 0 where low and required 2 to 4Ah of capacity added to each.

```python
import matplotlib.pyplot as plt

# Board indices and corresponding average voltages
boards = [0, 1, 2, 3, 4, 5, 6, 7]
avg_voltages = [4.0619, 4.0687, 4.0701, 4.0649, 4.0675, 4.0633, 4.0636, 4.0566]

plt.figure(figsize=(8, 4))
plt.plot(boards, avg_voltages, marker='o', linestyle='-', color='b')
plt.xlabel('Board Number')
plt.ylabel('Average Cell Voltage (V)')
plt.title('Average Cell Voltage per Board')
plt.xticks(boards)
plt.grid(True)
plt.tight_layout()
plt.show()
```
