Statistical Analysis & Report Summary
Voltage Drop Calculation:
The voltage drop for each of the 86 cells was calculated as the difference between the charged (Data Set 1) and uncharged (Data Set 2) cell voltages. This indicates the discharge each cell experienced during use.

Statistical Method:
We applied the z-score method to the voltage drops. The z-score standardizes the differences, allowing us to determine which cells exhibit unusually high or low voltage drops relative to the overall distribution.

Observations:

The mean voltage drop provides an average value for the discharge across cells, while the standard deviation quantifies the spread.
Most cells fall within a narrow range of z-scores (typically within ±2), indicating a consistent performance across the battery pack.
Cells with z-scores significantly outside this range (e.g., >2 or <–2) may indicate anomalies that warrant further investigation.
Summary:
The analysis reveals that the majority of the 86 cells exhibit similar voltage drops, with a consistent mean and relatively low standard deviation. The application of z-score analysis confirms that there are no extreme outliers, suggesting uniform cell performance. Any cells with z-scores beyond ±2 would be considered potential anomalies for further review.
