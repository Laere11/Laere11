```python
import matplotlib.pyplot as plt

# Data
mah = [0, 5725, 9650, 12200, 13335]
voltage = [3.34, 3.75, 3.95, 4.05, 4.08]

plt.figure(figsize=(8, 6))
plt.plot(mah, voltage, marker='o', linestyle='-', color='b')
plt.title("Enerdel Pak_A")
plt.xlabel("mAh in")
plt.ylabel("Cell Voltage (V)")
plt.grid(True)
plt.tight_layout()
plt.savefig("enerdel_plot.png")
plt.show()
```
